# numpy

## numpy 산술연산시 특징
- array의 길이가 같아야한다.
- 순서대로 계산을한다
    - ex) [1, 2, 3] + [2, 3, 4]
    - [2, 5, 7]
- 브로드캐스트 지원
    - ex) [1.0, 2.0, 3.0] / 2.0
    - [0.5, 1., 0.5]

## N차원 배열 가능
```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
print(A)
([1, 2]
  [3, 4])

A.Shape() # 형상
(2,2)
A.dtype() # 원소의 type
```

> 수학에서 1차원은 벡터, 2차원은 행렬/ 벡터와 행렬을 일반화 한 것을 텐서 라고한다. 3차원이상은 다차원배열이라 부른다

- Numpy는 형상이 다른 배열도 계산가능 -> 브로드캐스트

# matplotlib 
- 그래프를 그려주는 라이브러리
- 그래프를 그리려면 matplotlib의 pyplot 모듈을 사용한다.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# 데이터 준비
x = np.arange(0, 6, 0.1) # 0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
````

# pyplot 기능

- sin, cos 등 함수 그래프 그림
- 제목, 각 축의 이름(레이블) 표시등

````python
y2 = np.cos(x)
plt.plot(x, y, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title('sin & cos') # 제목
plt.legend()
plt.show()
````

# 이미지 표시하기
